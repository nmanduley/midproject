```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
# confirmed_global
df_conf = pd.read_csv('confirmed_global.csv') 
df_conf.head()
```

```python
# deaths_global
df_deaths = pd.read_csv('deaths_global.csv') 
df_deaths.head()
```

```python
# recovered_global
df_rec = pd.read_csv('recovered_global.csv') 
df_rec.head()
```

### Verificar NaN

```python
# Hay confirmed_global, hay un valor Nan en "Lat" y "Long" correspondiente a Canada / Repatriated Travellers
df_conf[df_conf["Long"].isnull()]
```

```python
# En deaths_global tambien (y en recovered_global no)
df_deaths[df_deaths["Long"].isnull()]
```

```python
# Como es una porcion minuscula de casos, esta fila sera suprimida
df_rec[df_rec["Lat"].isnull()]
```

```python
# En recovered_global hay una sola entrada para Canada
df_rec[df_rec["Country/Region"] == "Canada"]
```

```python
# NaN por columna:
df_conf.isnull().sum()
# En el preview aparecen 189 valores NaN en la columna Province/State, y 1 en Lat y Long.
# Haciendo .sum() de lo anterior, el total de NaN en el dataset da 191. Por lo tanto, no hay NaN en las columnas de fechas.
# Ocurre algo similar con los tres datasets. Por lo tanto, solo hace falta lidiar con los NaN de la columna de provincias, y los de Lat y Long.
```

### Para mayor manejabilidad, los paises cuyas estadisticas estÃ©n especificadas por provincia seran integrados en una sola fila, y la columna "Province/State" sera desechada

```python
# Confirmed
grouped_conf = df_conf.groupby('Country/Region').sum().reset_index()

# Deaths
grouped_deaths = df_deaths.groupby('Country/Region').sum().reset_index()

# Recovered
grouped_rec = df_rec.groupby('Country/Region').sum().reset_index()
```

### Exportar nuevos CSV

```python
grouped_conf.to_csv('grouped_confirmed.csv', index=False)
grouped_deaths.to_csv('grouped_deaths.csv', index=False)
grouped_rec.to_csv('grouped_recovered.csv', index=False)
```

```python
country = "Spain"
data = df_conf[df_conf["Country/Region"]==country]

xvals = [datetime.strptime(col_name, '%m/%d/%y') for col_name in data.columns[4:]]

fig = plt.figure(figsize=(10,5))
plt.plot(xvals, data.iloc[0,4:])
plt.grid('minor')
plt.title(f"Total confirmed deaths in {country}")
```

```python
countries = ["Spain", "Panama", "Venezuela", "Colombia"]

fig = plt.figure(figsize=(10,5))
for country in countries:
    data = df_conf[df_conf["Country/Region"]==country]
    xvals = [datetime.strptime(col_name, '%m/%d/%y') for col_name in data.columns[4:]]
    
    plt.plot(xvals, data.iloc[0,4:])
    plt.grid('minor')

    country_lst = ', '.join(countries)
plt.title(f"Total confirmed deaths in {country_lst}")
plt.legend(countries)
```
